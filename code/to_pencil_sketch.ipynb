```python
import cv2
from sketch import pencil_sketch_from_bgr, save_image
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../datasets/dog.jpg")

results = pencil_sketch_from_bgr(
    img,
    ksize=21,               # Gaussian blur kernel size: controls overall smoothness (larger = softer sketch)
    edge_strength=0.1,      # Edge reinforcement strength: enhances main contours without over-emphasizing outlines
    texture_strength=0.55,  # Texture (fur) strength: controls visibility of fine hair-like textures
    detail_sigma=1.6,       # Texture scale: smaller values capture finer hair details, larger values produce coarser texture
    contrast=1.05,          # Contrast adjustment: slightly darkens strokes while preserving highlights
    gamma=1.0,              # Gamma correction: 1.0 keeps original tonal distribution
    darken=0,               # Global darkening: 0 means no additional darkening, keeps background clean
)



save_image(
    path="../outputs/dog_sketch.png",
    img=results["sketch_final"]
)

```

```python
results = pencil_sketch_from_bgr(
    img,
    ksize=21,              
    edge_strength=0.05,    
    texture_strength=0.7,  
    detail_sigma=3.8,  
    contrast=1.0,
    gamma=0.9,
    darken=8,
)
save_image(
    path="../outputs/dog_sketch.png",
    img=results["sketch_final"]
)
```
